```python
#importing basic libraries
import warnings
warnings.filterwarnings('ignore')
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import string
import spacy
from matplotlib.pyplot import imread
from wordcloud import WordCloud
%matplotlib inline
```

```python
Data=pd.read_csv('Elon_musk.csv',encoding='latin1')
Data=Data.drop(['Unnamed: 0'],axis=1)
Data
```

# Text Mining

```python
Data=[Text.strip() for Text in Data.Text]
Data=[Text for Text in Data if Text]
Data
```

```python
# Joining the list into one string 
Text=''.join(Data)
Text
```

```python
#Punctuation
no_punc_Text = Text.translate(str.maketrans('','', string.punctuation))
no_punc_Text
```

# Tokenization

```python
#Importing Natural Language tool kit
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
text_Tokens=word_tokenize(no_punc_Text)
text_Tokens 
```

```python
len(text_Tokens)
```

```python
#Removing Stopwords with help of nltk
from nltk.corpus import stopwords
my_stop_words=stopwords.words('english')
sw_list = pd.read_csv('stop.txt')
my_stop_words.extend(sw_list)
no_stop_tokens=[word for word in text_Tokens if not word in my_stop_words]
print(no_stop_tokens)
```

```python
# Normalize the data (transforming all charecters into lower case)
lower_words=[Text.lower() for Text in no_stop_tokens]
print(lower_words[0:200])
```

# stemming

```python
# Stemming 
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_tokens=[ps.stem(word) for word in lower_words]
print(stemmed_tokens[0:200])
```

```python
len(stemmed_tokens)
```

# Lemmatization

```python
# Lemmatization
nlp=spacy.load('en_core_web_sm')
doc=nlp(' '.join(lower_words))
print(doc)
```

```python
lemmas=[token.lemma_ for token in doc]
print(lemmas)
```

```python
clean_tweets=' '.join(lemmas)
clean_tweets
```

# Feature Extaction

```python
# Filtering the nouns and verbs only
nouns_verbs=[token.text for token in doc_block if token.pos_ in ('NOUN','VERB')]
print(nouns_verbs[0:200])
```

```python
# Counting the noun & verb tokens
#from sklearn.feature_extraction.text import CountVectorizer
#cv=CountVectorizer()

X=cv.fit_transform(nouns_verbs)
sum_words=X.sum(axis=0)
words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key=lambda x: x[1], reverse=True)
wf_df=pd.DataFrame(words_freq)
wf_df.columns=['word','count']
wf_df[0:10] # viewing top ten results
```
